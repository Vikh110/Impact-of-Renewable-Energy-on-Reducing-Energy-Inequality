```python
# Standardize GeoAreaName in all datasets
for df in [df1, df2, df3, df4]:
    df['GeoAreaName'] = df['GeoAreaName'].str.strip()  # Remove extra spaces
    df['GeoAreaName'] = df['GeoAreaName'].str.title()  # Standardize case

```

```python
# Define regions and countries of interest
regions_of_interest = [
    'Eastern Asia', 'Europe','South-Eastern Asia', 'Southern Asia',
    'Sub-Saharan Africa', 'Western Asia', 'World'
]

countries_of_interest = ['India', 'China']

# Filter each dataset
df1_filtered = df1[df1['GeoAreaName'].isin(regions_of_interest + countries_of_interest)]
df2_filtered = df2[df2['GeoAreaName'].isin(regions_of_interest + countries_of_interest)]
df3_filtered = df3[df3['GeoAreaName'].isin(regions_of_interest + countries_of_interest)]
df4_filtered = df4[df4['GeoAreaName'].isin(regions_of_interest + countries_of_interest)]
```

```python
# Merge the filtered datasets using outer join
merged_df = df1_filtered.merge(df2_filtered, on=['GeoAreaName', 'TimePeriod'], how='outer')
merged_df = merged_df.merge(df3_filtered, on=['GeoAreaName', 'TimePeriod','Type of renewable technology'], how='outer')
final_df = merged_df.merge(df4_filtered, on=['GeoAreaName', 'TimePeriod'], how='outer')

# Inspect the merged dataset
print("Final dataset shape:", final_df.shape)
print(final_df.head())
```

```python
# Check missing values
print(final_df.isnull().sum())

# Inspect rows with missing values
print(final_df[final_df.isnull().any(axis=1)])
```

```python
final_df
```

```python
final_df['Type of renewable technology'].fillna('Unknown', inplace=True)
```

```python
final_df
```

```python
# Fill missing values in Intl_Financial_Flows_Clean_Energy_USD with regional median
final_df['Intl_Financial_Flows_Clean_Energy_USD'] = final_df.groupby('GeoAreaName')[
    'Intl_Financial_Flows_Clean_Energy_USD'].transform(
    lambda x: x.fillna(x.median())
)

# Fill remaining NaNs with global median
final_df['Intl_Financial_Flows_Clean_Energy_USD'].fillna(
    final_df['Intl_Financial_Flows_Clean_Energy_USD'].median(), inplace=True
)
```

```python
print(final_df.isnull().sum())
```

```python
# Step 1: Interpolate missing values within each region (linear interpolation over time)
final_df['Renewable_Share_TFEC'] = final_df.groupby('GeoAreaName')['Renewable_Share_TFEC'].transform(
    lambda x: x.interpolate(method='linear', limit_direction='both')
)

# Step 2: Fill remaining NaNs with regional median
final_df['Renewable_Share_TFEC'] = final_df.groupby('GeoAreaName')['Renewable_Share_TFEC'].transform(
    lambda x: x.fillna(x.median())
)

# Step 3: Fill remaining NaNs with global median (for regions with no data at all)
final_df['Renewable_Share_TFEC'].fillna(
    final_df['Renewable_Share_TFEC'].median(), inplace=True
)
```

```python
# Fill missing values in the Location column with "Unknown"
final_df['Location'] = final_df['Location'].fillna('Unknown')
```

```python
# Fill missing Access_to_Electricity_Percent with regional median
final_df['Access_to_Electricity_Percent'] = final_df.groupby('GeoAreaName')[
    'Access_to_Electricity_Percent'
].transform(lambda x: x.fillna(x.median()))
```

```python
# Fill missing Renewable_Capacity_Watts_Per_Capita with regional mean
final_df['Renewable_Capacity_Watts_Per_Capita'] = final_df.groupby('GeoAreaName')[
    'Renewable_Capacity_Watts_Per_Capita'
].transform(lambda x: x.fillna(x.mean()))
```

```python
# Check for remaining missing values
print(final_df.isnull().sum())
```

* Filling Location with "Unknown" avoids introducing bias and ensures these rows remain usable for analysis.
* Using median for Access_to_Electricity_Percent is robust to outliers and maintains the region-specific trends.
* Using mean for Renewable_Capacity_Watts_Per_Capita smoothens gaps without drastic distortions.

```python
# Rename columns for clarity
final_df.rename(
    columns={
        'GeoAreaName': 'Region',
        'TimePeriod': 'Year',
        'Access_to_Electricity_Percent': 'Electricity_Access_Percentage',
        'Location': 'Area_Type',
        'Intl_Financial_Flows_Clean_Energy_USD': 'Financial_Flows_Clean_Energy_USD',
        'Type of renewable technology': 'Renewable_Technology_Type',
        'Renewable_Capacity_Watts_Per_Capita': 'Renewable_Capacity_Per_Capita_Watts',
        'Renewable_Share_TFEC': 'Renewable_Share_in_TFEC_Percentage'
    },
    inplace=True
)
```

The frequency on the y-axis in the histograms represents the number of data points (regions, years, or entries in your dataset) that fall within a particular range (or bin) of the x-axis variable.

Explanation:

Bins on the X-Axis:

The x-axis is divided into intervals or ranges (bins). For example, if the x-axis represents "Electricity Access Percentage," a bin might represent the range 70–80%.

Frequency on the Y-Axis:

The y-axis indicates how many data points fall within each bin.

For example:

If one bin is 70–80% electricity access, and the corresponding frequency is 500, it means 500 entries in the dataset have electricity access values between 70% and 80%.

Visual Example:

In the histogram for Electricity Access Percentage, you see:

A high frequency near the bin for 100% electricity access, meaning most data points (regions or years) report nearly full electricity access.

Lower frequencies for bins below 50%, showing fewer regions have low electricity access.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Filter the dataset for "Rural" and "Urban"
filtered_data = final_df[final_df["Area_Type"].isin(["Rural", "Urban"])]

# Unique regions
regions = filtered_data["Region"].unique()

# Iterate through each region and create plots
for region in regions:
    region_data = filtered_data[filtered_data["Region"] == region]
    
    # Bar Plot: Rural vs Urban Access to Electricity by Year
    plt.figure(figsize=(8, 6))
    sns.barplot(
        data=region_data,
        x="Year",
        y="Electricity_Access_Percentage",
        hue="Area_Type",
        palette="Set2"
    )
    plt.title(f"Rural vs Urban Electricity Access Percentage in {region}", fontsize=14)
    plt.xlabel("Year", fontsize=12)
    plt.ylabel("Access to Electricity Percentage", fontsize=12)
    plt.legend(title="Area Type", fontsize=10)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
    
    # KDE Plot: Rural vs Urban Density of Electricity Access
    plt.figure(figsize=(8, 6))
    sns.kdeplot(
        data=region_data[region_data["Area_Type"] == "Rural"],
        x="Electricity_Access_Percentage",
        label="Rural",
        shade=True,
        color="blue"
    )
    sns.kdeplot(
        data=region_data[region_data["Area_Type"] == "Urban"],
        x="Electricity_Access_Percentage",
        label="Urban",
        shade=True,
        color="orange"
    )
    plt.title(f"Density of Electricity Access Percentage in {region}", fontsize=14)
    plt.xlabel("Access to Electricity Percentage", fontsize=12)
    plt.ylabel("Density", fontsize=12)
    plt.legend(fontsize=10)
    plt.tight_layout()
    plt.show()
```
